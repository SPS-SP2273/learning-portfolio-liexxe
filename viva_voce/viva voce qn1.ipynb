```python
def get_genome1(person_id):
    filename=f'person_{person_id:05d}.txt'
    with open(filename) as file:
        file_content=file.read()
    return file_content
```

```python
#2
list_of_id=[1,2,3]

def pool(list_of_id): #alternatively pool(*list_of_id), * = unpaacks,
                        #but not need as for i unpacked already
    mix = ''
    for i in list_of_id: #i=1,2,3
        mix += get_genome1(i)
    return mix
```

```python
#testing groups of 10
test_count=0
for i in range(0,100,10):
    test_count+=1
    list_of_id = [*range(i,i+10)] #makes list 
    if test(pool(list_of_id)) == True:
        print(f'Test for group {test_count} is POSITIVE')
    elif test(pool(list_of_id)) == False:
        print(f'Test for group {test_count} is NEGATIVE')
    #for true test, need further testing to find who, split into twos
```

```python
#groups of 10 followed by testing individuals
test_count=0
group_count = 0
for i in range(0,100,10):
    test_count+=1
    group_count+=1
    list_of_id = [*range(i,i+10)] #makes list
    
    if test(pool(list_of_id)) == True:
        print(f'Test {test_count}, group {group_count} is POSITIVE')
        
        for indiv in list_of_id:
            test_count+=1
            if test(pool([indiv])) == True: #pool only accepts list
                  print(f'Test{test_count}, Individual {indiv} is POSITIVE')
            elif test(pool([indiv])) == False:
                  print(f'Test{test_count}, Individual {indiv} is NEGATIVE')

    elif test(pool(list_of_id)) == False:
        print(f'Test {test_count}, group {group_count} is NEGATIVE')

print(f'Number of test done is {test_count}')
```
